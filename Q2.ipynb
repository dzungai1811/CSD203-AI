```python
class EmptyQueueException(Exception):
    pass

class QueueNode:
    def __init__(self, value):
        self.value = value
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def isEmpty(self):
        return self.front is None

    def clear(self):
        self.front = None
        self.rear = None

    def enqueue(self, x):
        new_node = QueueNode(x)
        if self.isEmpty():
            self.front = new_node
            self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node

    def dequeue(self):
        if self.isEmpty():
            raise EmptyQueueException("Queue is empty")
        value = self.front.value
        self.front = self.front.next
        return value

    def first(self):
        if self.isEmpty():
            raise EmptyQueueException("Queue is empty")
        return self.front.value

    def traverse(self):
        current = self.front
        while current:
            print(current.value, end=" ")
            current = current.next
        print()

def real_to_binary(real):
    queue = Queue()
    precision = 8
    while real > 0 and precision > 0:
        real *= 2
        bit = int(real)
        queue.enqueue(bit)
        real -= bit
        precision -= 1

    binary = ""
    while not queue.isEmpty():
        binary += str(queue.dequeue())

    return binary
```
